## Numpy
Numpy allows us to manipulate arrays

```python
import numpy

# the following is an initialzation of an array
mylist = [[1, 2, 3], [3, 4, 5]]
myarray = numpy.array(mylist)
```

```python
# this return the dimension of the array
myarray.shape
```

```python
# print row of an array
myarray[1]
```

```python
# print column of an array 
myarray[:, 2]
```

## Pandas

```python
import pandas as pd

# initialization of Series
my_list = [1, 2, 3, 1]
my_row_names = ["Ann", "Ben", "Clay", "Alice"]
my_series = pd.Series(my_list, my_row_names)
```

```python
my_series
```

```python
my_series[1]
```

```python
# shows the indices and values
print(my_series.index)
print(my_series.values)
```

```python
# series can be used as dictionary
my_series["Ann"]
```

```python
my_series["Ann":]
```
